```python
plt.figure(figsize=(5,6))
plt.title("Reservation Status Count")
plt.ylabel("Count")
plt.bar(['Not Cancelled'],data['is_canceled'].value_counts(),edgecolor='k')
```

```python
snb.countplot(x='hotel',hue='is_canceled',data=data)
plt.legend(["Not Cancelled","Cancelled"])
```

```python
CityHotel=data[data['hotel']=='City Hotel']
CityHotel['is_canceled'].value_counts(normalize=True)*100
```

```python
ResortHotel=data[data['hotel']=='Resort Hotel']
ResortHotel['is_canceled'].value_counts(normalize=True)*100
```

```python
# data=data[['reservation_status_date','adr']]
# data.head()
```

```python
CityHotel=CityHotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
ResortHotel=ResortHotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.plot(CityHotel.index,CityHotel['adr'],label='CityHotel')
plt.plot(ResortHotel.index,ResortHotel['adr'],label='ResortHotel')
plt.legend(fontsize=20)
```

```python
data['month']=data['reservation_status_date'].dt.month
fig=plt.figure(figsize=(18,7))
snb.countplot(x = 'month', hue='is_canceled', data= data, palette='bright')
plt.title("Reservation Status per Month",size=20)
plt.xlabel("Months",size=12)
plt.ylabel("Num of Reservations",size=12)
plt.legend(['Not_canceled','canceled'])
plt.savefig("Reservation_Status_per_Month.png")
plt.show()
```

```python
df=data
plt.figure(figsize=(14,5))
plt.title("ADR per month")
ax11=snb.barplot(x= 'month' , y= 'adr',data = df[df['is_canceled']==1].groupby("month")[['adr']].sum().reset_index())
plt.xlabel("Month")
plt.ylabel("ADR")
plt.show()
```

```python
canceled_=df[df['is_canceled']==1]
top_ten_country=canceled_["country"].value_counts()[:10]
```

```python
plt.figure(figsize=(10,5))
plt.title("Top 10 Countries with max canellations")
plt.pie(top_ten_country, autopct='%.1f',labels = top_ten_country.index)
```
