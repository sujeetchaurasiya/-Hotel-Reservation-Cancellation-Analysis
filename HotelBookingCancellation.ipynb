```python
not_canceled=df[df["is_canceled"]==0]
```

```python
canceled_bookings_adr=canceled_.groupby("reservation_status_date")[["adr"]].mean()
canceled_bookings_adr.reset_index(inplace=True)
canceled_bookings_adr.sort_values("reservation_status_date",inplace=True)

not_canceled_adr=not_canceled.groupby("reservation_status_date")[['adr']].mean()
not_canceled_adr.reset_index(inplace=True)
not_canceled_adr.sort_values("reservation_status_date", inplace=True)

```

```python
not_canceled_adr=not_canceled_adr[(not_canceled_adr["reservation_status_date"]>'2016') & (not_canceled_adr["reservation_status_date"]<='2017')]
canceled_bookings_adr=canceled_bookings_adr[(canceled_bookings_adr["reservation_status_date"]>"2016") & (canceled_bookings_adr["reservation_status_date"]<="2017")]
```

```python
plt.figure(figsize=(16,5))
plt.plot(canceled_bookings_adr["reservation_status_date"],canceled_bookings_adr["adr"],label="Canceled")
plt.plot(not_canceled_adr["reservation_status_date"],not_canceled_adr["adr"],label=" Not Canceled")
plt.title("ADR rate for canceled and not canceled Reservations", size= 18)
plt.xlabel("Date", size=14)
plt.ylabel("ADR", size=14 )
plt.legend()
plt.savefig("ADR rate for canceled and not canceled Reservations")
```
